```python
import numpy as np
import csv
import pandas as pd
```

```python
#Creating a function to help us collect data and arrange them is the right format
def collectData(df,year,serie):
    j=year-1996
    country = []
    countrySerie = []
    years=[]
    df_serie=pd.DataFrame(columns=['Country', 'Serie'])
    for i in range(len(df)):
        if df.iloc[i][2] not in country:
            country.append(df.iloc[i][2])
        if df.iloc[i][1]==serie:
            countrySerie.append(df.iloc[i][j])
            years.append(year)
    df_serie['Country']=country
    df_serie['Serie']=countrySerie
    df_serie['Year']=years
    return df_serie
```

```python
#Loading the initial datasets and series names
raw_df=pd.read_csv('Dataset.csv')

series=['SP.POP.TOTL','EN.POP.DNST','SI.POV.DDAY','SI.POV.UMIC','SI.POV.GINI','SP.DYN.LE00.IN','NY.GDP.PCAP.KD',
        'SE.ADT.LITR.ZS','SP.DYN.TFRT.IN','EN.URB.MCTY.TL.ZS','SP.RUR.TOTL.ZS','SL.UEM.TOTL.ZS','SP.POP.0014.TO.ZS',
            'SP.POP.1564.TO.ZS','SP.POP.65UP.TO.ZS']
seriesName=['Population','Density','Poverty190','Poverty590','GiniIndex',
           'LifeExp','GDPCapita','LiteracyRate','FertilityRate','PopUrban1M','RuralPop','Unemployment',
               'Pop0-14','Pop15-64','Pop65']
raw_df
```

```python
#Collect the first serie and create the dataframe
serie0='SP.POP.TOTL'
df1=collectData(raw_df,2000,serie0)
for y in range(2001,2020):
    df2=collectData(raw_df,y,serie0) 
    frames=[df1,df2]
    df1=pd.concat(frames,axis=0)
df1=df1.rename(columns={'Serie':'Population'})
print(df1)
```

```python
#Loop on series to collect all values
for i in range(0,len(series)):
    df3=collectData(raw_df,2000,series[i])
    for y in range(2001,2020):
        df4=collectData(raw_df,y,series[i])
        frames=[df3,df4]
        df3=pd.concat(frames,axis=0)
    df3=df3.rename(columns={'Serie': seriesName[i]})
    df3=df3.drop(['Year','Country'],axis=1)
    df1=pd.concat([df1,df3],axis=1)
```
